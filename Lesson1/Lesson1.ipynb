# Lesson 1: Introduction to Financial Big Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
# Define the ticker symbol
ticker_symbol = 'AAPL'

# Get data for the ticker
ticker_data = yf.Ticker(ticker_symbol)

# Get historical market data
ticker_df = ticker_data.history(period='1d', start='2001-1-1', end='2023-1-1')

# Display the data
print(ticker_df)
```

```python
# Plot the log returns of the closing price
plt.figure(figsize=(10, 5))
plt.plot(np.log(ticker_df['Close'] / ticker_df['Close'].shift(1)))
plt.title('Log Returns of ' + ticker_symbol)
plt.show()
```
